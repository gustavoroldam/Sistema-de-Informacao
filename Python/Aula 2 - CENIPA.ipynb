```python
df_view = df_ocorrencia.value_counts(subset=['ocorrencia_ano']).sort_index(ascending=True).to_frame().reset_index() \
    .rename(columns={'ocorrencia_ano': 'ano', 'count': 'total'})

#Diferença
df_view['total-diff'] = df_view['total'].diff().fillna(0).astype(int)

#Diferença em %
df_view['total-pct-change'] = df_view['total'].pct_change().fillna(0).round(2)
df_view['total-pct-change-fmt'] = (df_view['total'].pct_change().fillna(0) * 100).round(2).astype(str) + '%'

df_view['total-sum-3anos'] = df_view['total'].rolling(window=3).sum().fillna(0).astype(int)
df_view['total-mean-3anos'] = df_view['total'].rolling(window=3).mean().fillna(0).astype(int) # Média

#Soma acimulada/cumulativo
df_view['total-cumsum'] = df_view['total'].cumsum()

df_view = df_view.sort_values(by=['ano'], ascending=[False], ignore_index=True)

df_view
```

```python
def get_lat_long_from(municipio, uf):
    municipio = unidecode(municipio)
    try:
        municipios_info = municipios_dict[(municipio, uf)]
        return (municipios_info['latitude'], municipios_info['longitude'])
    except:
        return ('-', '-')
    
print(get_lat_long_from('ASSIS', 'SP'))
print(get_lat_long_from('BRASÍLIA', 'DF'))
print(get_lat_long_from('X', 'Y'))
```

```python
import folium
from folium.plugins import MiniMap, Fullscreen, LocateControl, Geocoder, Draw, MeasureControl, MousePosition, TagFilterButton

m = folium.Map(location=None,
               zoom_start=3, min_zoom=3, max_zoom=20,
               tiles=None,
               prefer_canvas=True, width='100%', height='100%')

#Adiciona o Tema ao mapa
folium.raster_layers.TileLayer('CartoDB Positron').add_to(m)
folium.raster_layers.TileLayer('CartoDB Voyager').add_to(m)
folium.raster_layers.TileLayer('OpenStreetMap').add_to(m)



MiniMap().add_to(m)

Fullscreen().add_to(m)

LocateControl().add_to(m)

Geocoder().add_to(m)

Draw().add_to(m)

MeasureControl().add_to(m)

MousePosition().add_to(m)

#Libera a opção de selecionar o tema depois da execução
TagFilterButton(['ACIDENTE', 'INCIDENTE', 'INCIDENTE GRAVE']).add_to(m)

markers = folium.FeatureGroup(name='Ocorrencias CENIPA (2023)', control=True).add_to(m)
circles = folium.FeatureGroup(name='Total Ocorrências CENIPA', control=True).add_to(m)

classificacao_dict = {
    'ACIDENTE': {'cor': 'blue', 'icone': 'bell'},
    'INCIDENTE': {'cor': 'orange', 'icone': 'exclamation'},
    'INCIDENTE GRAVE': {'cor': 'red', 'icone': 'warning'}
}

def get_info_html(geo_item):
    info_html = f'<div style="white-space: normal">'
    info_html += f'<img src="https://github.com/bgeneto/bandeiras-br/blob/master/imagens/{geo_item["ocorrencia_uf"]}.png?raw=true" width="40%" height="40%" /> <br>'
    for key, value in geo_item.items():
        info_html += f'<b>{key}</b>: {value} <br/>'
    info_html += f'</div>'
    return folium.Html(info_html, script=True, width=300).render()

for ocorrencias_geo_item in ocorrencias_geo_dict[:100]:
    cor_ocorrencia = classificacao_dict[ocorrencias_geo_item['ocorrencia_classificacao']]['cor']
    icone_ocorrencia = classificacao_dict[ocorrencias_geo_item['ocorrencia_classificacao']]['icone']
    
    marker_html = get_info_html(ocorrencias_geo_item)
    
    marker = folium.Marker(location=[ocorrencias_geo_item['ocorrencia_latitude'], ocorrencias_geo_item['ocorrencia_longitude']],
                          icon=folium.Icon(color=cor_ocorrencia, prefix='fa', icon=icone_ocorrencia),
                          tags=[ocorrencias_geo_item['ocorrencia_classificacao']],
                          popup=folium.Popup(html=marker_html),
                          tooltip=folium.features.Tooltip(text=marker_html, sticky=False))
    markers.add_child(marker)
    
for ocorrencia_total_item in ocorrencias_total_dict:
    cor_circle = custom_colormap(ocorrencia_total_item['count'])
    circle_html = get_info_html(ocorrencia_total_item)
    
    circle = folium.Circle(location=[ocorrencia_total_item['ocorrencia_latitude'],
                                    ocorrencia_total_item['ocorrencia_longitude']],
                          radius=ocorrencia_total_item['count'] * 500,
                          fill=True, fill_color=cor_circle, color=cor_circle,
                          popup=folium.Popup(html=circle_html),
                          tooltip=folium.features.Tooltip(text=circle_html, sticky=False))
    circles.add_child(circle)

folium.LayerControl(collapsed=False).add_to(m)

m.fit_bounds(markers.get_bounds())

m.save(outfile='mapa_cenipa.html')

m
```
