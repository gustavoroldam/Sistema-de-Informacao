```python
import pandas as pd
import re # re = expressao regulares
import urllib.request
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df_view = df.copy()
df_view
```

```python
df_view = df_view.rename(columns={'Date': 'Data', 'Workout Name': 'Nome do Treino', 'Duration': 'DuraÃ§Ã£o', 'Exercise Name': 'Nome do Exercicio', \
                                 'Set Order': 'Sets', 'Weight': 'Peso', 'Reps': 'Repeticoes', 'Distance': 'Distancia', 'Seconds': 'Segundos',  \
                                 'Notes': 'Notas', 'Workout Notes': 'AnotaÃ§Ãµes do Treino'})
df_view
```

```python
df_view .value_counts(subset=['Nome do Treino', 'Nome do Exercicio']).to_frame('RPE').reset_index() \
    .sort_values(by=['RPE'], ascending=[True], ignore_index=True)
```

```python
df_view .value_counts(subset=['Nome do Treino', 'Nome do Exercicio']).to_frame('RPE').reset_index() \
    .sort_values(by=['RPE'], ascending=[True], ignore_index=True).query('RPE > 100')
```

```python
df_view.query('RPE >= 1') \
    .groupby(by=['Nome do Treino', 'Nome do Exercicio', 'Sets']) \
    .agg({'RPE': ['min', 'max', 'mean']}).reset_index()
```

```python
df_view.query('RPE >= 1').groupby(by= ['Nome do Treino', 'Nome do Exercicio', 'Sets']).agg({'RPE' : ['min','max','mean']}).round(2).reset_index()
```
