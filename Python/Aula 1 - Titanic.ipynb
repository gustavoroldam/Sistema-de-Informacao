```python
def show_na_columns(df, total_column_name='total', hide_zero_values=True):
    df_view = df.isna().sum().to_frame().reset_index().rename(columns={'index': 'column', 0: total_column_name}) \
        .sort_values(by=[total_column_name, 'column'], ascending=[False, True], ignore_index=True).copy()
    if hide_zero_values:
        df_view = df_view.query(f'{total_column_name} > 0').copy()
    return df_view
```
